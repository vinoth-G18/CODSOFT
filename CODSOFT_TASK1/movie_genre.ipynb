TEXT PREPROCESSING AND REPRESENTATION

```python
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import word_tokenize
import re
```

```python
lemmatizer=WordNetLemmatizer()
```

```python
stop_words=set(stopwords.words("english"))
```

```python
def preprocessing(text):
    text = re.sub(r'\d+',' ', text)
    text=text.lower()
    tokens=word_tokenize(text)
    filtered_tokens=[word for word in tokens if word.isalnum() and word not in stop_words]
    lemmatized_tokens=[lemmatizer.lemmatize(word) for word in filtered_tokens]
    return " ".join(lemmatized_tokens)
```

```python
X_train=X_train['Description'].astype(str).apply(preprocessing)
```

```python
X_test=X_test['Description'].astype(str).apply(preprocessing)
```

```python
y_test=test_data_solution.drop(columns=['title','Description'])
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfid=TfidfVectorizer(ngram_range=(1, 2),max_features=10000)
```

```python
X_train=tfid.fit_transform(X_train)
```

```python
X_test=tfid.transform(X_test)
```
